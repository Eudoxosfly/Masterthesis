```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Load the data
df = pd.read_csv('volumes.csv', names=['volume_mm3'])
```

```python
# Calculate the volume in mm^3
df["diameter_um"] = df["volume_mm3"].apply(lambda x: np.cbrt(x))
```

```python
# discard small volumes
thresh = 20
df = df[df["diameter_um"] > thresh]
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.histplot(df, x="diameter_um", bins=100, ax=ax, kde=True)
ax.set_xlabel('Diameter in um')

ax.set_ylabel('Count')

plt.show()
```
