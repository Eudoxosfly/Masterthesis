```python
from skimage import io
from mt.visualization import export_image
```

# Image export
This script is used to export SEM and optical microscopy images with a scale bar. The scale is based on the original image's scale bar.

```python
scales = {"B6_SEM_particle": {"scale_mm": 20 / 270 * 1e-3, "fixed_length_type": "single_particle"},
          "B2_SEM_overview": {"scale_mm": 100 / 150 * 1e-3, "fixed_length_type": "particles"},
          "B2_SEM_particle": {"scale_mm": 20 / 114 * 1e-3, "fixed_length_type": "single_particle"},
          "B2_SEM_detail": {"scale_mm": 3 / 95 * 1e-3, "fixed_length_type": "detail"},
          "B2_LM": {"scale_mm": 100 / 137 * 1e-3, "fixed_length_type": "overview"},
          "B6_LM": {"scale_mm": 100 / 137 * 1e-3, "fixed_length_type": "overview"},
          "B9_LM": {"scale_mm": 100 / 137 * 1e-3, "fixed_length_type": "overview"},
          "B25_LM": {"scale_mm": 100 / 137 * 1e-3, "fixed_length_type": "overview"},
          "B6_SEM_overview": {"scale_mm": 200 / 148 * 1e-3, "fixed_length_type": "particles"},
          "B6_SEM_detail": {"scale_mm": 2 / 184 * 1e-3, "fixed_length_type": "detail"},
          "B2_raw_LM": {"scale_mm": 100 / 137 * 1e-3, "fixed_length_type": "overview"},
          }

for sample in scales.keys():
    path = "../../05_Thesis/figures/Batches/" + sample + ".png"
    im = io.imread(path)
    export_image(im=im,
                 scale_mm=scales[sample]["scale_mm"],
                 file_path=path,
                 fixed_length_type=scales[sample]["fixed_length_type"])
```
